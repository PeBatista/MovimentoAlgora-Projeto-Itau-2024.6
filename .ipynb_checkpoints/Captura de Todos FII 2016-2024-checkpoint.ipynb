```python
import numpy as np
```

```python
# Função para remover o símbolo de R$ e converter em float
def remover_cifrao(valor):
    if isinstance(valor, str) and 'R$' in valor:
        return float(valor.replace('R$', '').replace('.', '').replace(',', '.'))
    return np.nan

# Função para converter porcentagem em decimal e aplicar sobre o valor base
def calcular_valor_decimal(vl_valor, vl_vpa):
    # Converta vl_vpa para um número, ou 0 se for nulo
    vl_vpa = remover_cifrao(vl_vpa)
    if pd.isna(vl_vpa):
        vl_vpa = 0
    
    if isinstance(vl_valor, str) and '%' in vl_valor:
        # Convertendo a variação percentual para decimal
        variacao_decimal = float(vl_valor.replace('%', '').replace(',', '.')) / 100
        return vl_vpa * (1 + variacao_decimal)
    elif isinstance(vl_valor, str) and 'R$' in vl_valor:
        # Se já for um valor monetário, apenas converter
        return remover_cifrao(vl_valor)
    return vl_vpa

# Aplicar a função de conversão na coluna 'vl_vpa'
algora['vl_vpa'] = algora['vl_vpa'].apply(remover_cifrao)

# Aplicar a função de cálculo para obter valores decimais baseados na variação
algora['valor_decimal'] = algora.apply(lambda row: calcular_valor_decimal(row['vl_valor'], row['vl_vpa']), axis=1)

# Exibir o DataFrame resultante
print(algora[['vl_valor', 'vl_vpa', 'valor_decimal']])
```
